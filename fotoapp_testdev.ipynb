```python
# Importar bibliotecas necesarias solo para el menu
import cv2
import fotoapp as fap
```

```python
def mostrar_menu():
    """
    Muestra el menú de opciones al usuario para la aplicación de procesamiento de imágenes.
    """
    print("\n--- Menú de Opciones para Procesamiento de Imágenes ---\n")
    print("1. Redimensionar imagen para redes sociales")
    print("2. Ajustar contraste de la imagen")
    print("3. Aplicar filtro a la imagen")
    print("4. Crear boceto a partir de la imagen")
    print("5. Salir")

```

```python

def main():
    """
    Función principal que ejecuta el programa de procesamiento de imágenes.
    Esta función principal gestiona un menú interactivo que permite al usuario realizar diversas operaciones
    sobre imágenes, como redimensionarlas para redes sociales, ajustar el contraste, aplicar filtros o crear
    un boceto a partir de la imagen. El programa seguirá ejecutándose hasta que el usuario elija la opción de
    salir (opción 5). Cada operación captura posibles errores y los muestra al usuario.

    """
    while True:
        mostrar_menu()
        opcion = input("\nSeleccione una opción: ")  # Solicitar la opción del usuario

        if opcion == '1':
            # Redimensionar imagen para redes sociales
            ruta_imagen = input("Ingrese la ruta de la imagen: ")
            print(ruta_imagen)
            plataforma = input("Ingrese la plataforma (YouTube, Instagram, Twitter o Facebook): ")
            try:
                redimensionada = fap.LeerAbrirRedimensionar(ruta_imagen, plataforma)
                print(f"Imagen redimensionada para {plataforma}.")
            except Exception as e:
                print(f"Error al redimensionar la imagen: {e}")

        elif opcion == '2':
            # Ajustar contraste de la imagen
            ruta_imagen = input("Ingrese la ruta de la imagen para ajustar el contraste: ")
            try:
                img = cv2.imread(ruta_imagen)
                fap.AjustarContrasteImg(img)
                print("Ajuste de contraste realizado con éxito.")
            except Exception as e:
                print(f"Error al ajustar el contraste de la imagen: {e}")

        elif opcion == '3':
            # Aplicar filtro a la imagen
            ruta_imagen = input("Ingrese la ruta de la imagen: ")
            print("Filtros disponibles para aplicar:\n"
                  "ORIGINAL\n"
                  "BLUR\n"
                  "CONTOUR\n"
                  "DETAIL\n"
                  "EDGE ENHANCE\n"
                  "EDGE ENHANCE MORE\n"
                  "EMBOSS\n"
                  "FIND EDGES\n"
                  "SHARPEN\n"
                  "SMOOTH")

            filtro = input("Ingrese el filtro deseado: ")
            try:
                fap.filtrosMostrarGuardar(ruta_imagen, filtro)
                print(f"Filtro '{filtro}' aplicado y guardado correctamente.")
            except Exception as e:
                print(f"Error al aplicar el filtro: {e}")

        elif opcion == '4':
            # Crear boceto a partir de la imagen
            ruta_imagen = input("Ingrese la ruta de la imagen: ")
            persona = input("¿Es una imagen de una persona? (si/no): ").strip().lower() == 'si'
            try:
                fap.crearBoceto(ruta_imagen, persona)
                print("Boceto creado y guardado correctamente.")
            except Exception as e:
                print(f"Error al crear el boceto: {e}")

        elif opcion == '5':
            #sale del programa
            print("Saliendo del programa.")
            break

        else:
            print("Opción no válida. Por favor, elija una opción del 1 al 5.")


```

```python
# Ejecutar el programa principal
main()
```
